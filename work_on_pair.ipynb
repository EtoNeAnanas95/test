```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [25, 32, 34, 20]

plt.plot(x, y, color = "orange", marker = "o", markersize = 8, label = "Линейный графчик")
plt.legend()
plt.grid()
plt.title("ГРАФИК ЧЕРТЕЙ")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["январь", "февраль", "март", "апрель"]
y = [25, 32, 34, 20]

plt.plot(x, y, color = "orange", marker = "o", markersize = 8, label = "Линейный графчик")

plt.bar(x, y, color = "pink", label = "Столбчатая диаграмма")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(y, labels = x, autopct="%1.1f%%")
plt.legend()
plt.show()
```

```python
data = np.random.rand(100)

plt.hist(data, bins = 20, edgecolor = "black", color = "purple")
plt.show()
```

```python
pokemon = pd.read_csv("Pokemon.csv")

plt.plot(pokemon["#"].head(50), pokemon["Attack"].head(50), marker = "o", label = "атака")
plt.plot(pokemon["#"].head(50), pokemon["HP"].head(50), marker = "o", label = "Здоровье")
plt.plot(pokemon["#"].head(50), pokemon["Defense"].head(50), marker = "o", label = "Защита")
plt.show()
```

```python
plt.bar(pokemon["#"].head(50), pokemon["Attack"].head(50), label = "атака")
plt.bar(pokemon["#"].head(50), pokemon["HP"].head(50), label = "Здоровье")
plt.bar(pokemon["#"].head(50), pokemon["Defense"].head(50), label = "Защита", alpha = 0.5)
plt.legend()
plt.show()
```
